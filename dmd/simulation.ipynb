```python
import numpy as np

simu_size = 200

# 生成若干个通话时长的模拟数据，T ~ N(400, 20)
T = np.random.normal(400, 20, simu_size)

# 生成若干个长途通话比例的模拟数据， L ~ U(0.1, 0.5)
L = np.random.uniform(0.1, 0.5, simu_size)

# 移动世界的月成本函数
def x1(t, l):
    if (t<=400):
        return 39.99
    
    return 39.99 + (t-400) * 0.4

# 自由通信的月成本函数
def x2(t, l):
    return 35 + t*l*0.08

# 模拟移动世界的月成本
X1 = np.array([x1(t, l) for t, l in zip(T, L)])

# 模拟自由通信的月成本
X2 = np.array([x2(t, l) for t, l in zip(T, L)])

# 自由通信与移动世界的月成本差值
delta = X2 - X1

print(delta)
```

```python
import matplotlib.pyplot as plt

plt.hist(delta, bins=30)
plt.xlabel("x2 - x1")
plt.ylabel("count")
plt.title("Histogram of x2 - x1")
plt.show()
```

```python
delta_mean =  np.mean(delta)
print(delta_mean)
delta_std = np.std(delta)
print(delta_std)

print(delta[delta>0].size)
print(np.median(delta))
print(delta.min())
```
